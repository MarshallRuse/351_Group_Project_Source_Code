```
# Reduce the memory usage of the dataframes
def reduce_memory(df):
  '''
  reduce_memory takes as input a dataframe, and transforms each column to be memory efficient.
  For instance, converting 64-bit integers to 32, object types to categorical, etc. 
  '''
  
  for col in df:
    
    # Transform ids to integers
    if ('SK_ID' in col):
      df[col] = df[col].fillna(0).astype(np.int32)
    
    # Transform object types to categorical
    elif (df[col].dtype == 'object') and (df[col].nunique() < df.shape[0]):
      df[col] = df[col].astype('category')
    
    # Transform binary integers to booleans
    elif (list(df[col].unique()) == [1,0]):
      df[col] = df[col].astype(bool)
      
    # Transform default float64 to float32
    elif df[col].dtype == float:
      df[col] = df[col].astype(np.float32)
    
    # Transform default int64 to int32
    elif df[col].dtype == int:
      df[col] = df[col].astype(np.int32)
      
  return df

for df in dataFrames:
  df = reduce_memory(df)
```

```
# Create featuretools EntitySets out of the dataframes
entitySet = ft.EntitySet(id='applications')

entitySet = entitySet.entity_from_dataframe(entity_id='appMerged', dataframe=appMerged, index='SK_ID_CURR')

entitySet = entitySet.entity_from_dataframe(entity_id='bureau', dataframe=bureau, index='SK_ID_BUREAU')

entitySet = entitySet.entity_from_dataframe(entity_id='bureauBalance', dataframe=bureauBalance, 
                              make_index=True, index='bureauBalance_index')

entitySet = entitySet.entity_from_dataframe(entity_id='POSCash', dataframe=POSCash, 
                              make_index=True, index='POSCash_index')

entitySet = entitySet.entity_from_dataframe(entity_id='creditCard', dataframe=creditCard,
                              make_index=True, index='creditCard_index')

entitySet = entitySet.entity_from_dataframe(entity_id='prevApps', dataframe=prevApps, index='SK_ID_PREV')

entitySet = entitySet.entity_from_dataframe(entity_id='installmentsPayments', dataframe=installmentsPayments,
                              make_index=True, index='installmentsPayments_index')

```

```
# Define the featuretools relationships
# relationships are hierarchical, with a one-to-many relationship as one descends
# through the hierarchy.  Dataframes are related by the indices of the parents
# acting as foreign keys of their children

# appMerged -> bureau
rel_appMerged_bureau = ft.Relationship(entitySet['appMerged']['SK_ID_CURR'],
                                       entitySet['bureau']['SK_ID_CURR'])

# bureau -> bureauBalance
rel_bureau_bureauBalance = ft.Relationship(entitySet['bureau']['SK_ID_BUREAU'], 
                                           entitySet['bureauBalance']['SK_ID_BUREAU'])

# appMerged -> prevApps
rel_appMerged_prevApps = ft.Relationship(entitySet['appMerged']['SK_ID_CURR'], 
                                         entitySet['prevApps']['SK_ID_CURR'])

# prevApps -> POSCash, creditCards, and installmentsPayments
rel_prevApps_POSCash = ft.Relationship(entitySet['prevApps']['SK_ID_PREV'],
                                       entitySet['POSCash']['SK_ID_PREV'])

rel_prevApps_creditCard = ft.Relationship(entitySet['prevApps']['SK_ID_PREV'], 
                                          entitySet['creditCard']['SK_ID_PREV'])

rel_prevApps_installmentsPayments = ft.Relationship(entitySet['prevApps']['SK_ID_PREV'], 
                                                    entitySet['installmentsPayments']['SK_ID_PREV'])


# add the relationships to the EntitySet
entitySet = entitySet.add_relationships([rel_appMerged_bureau,
                                        rel_bureau_bureauBalance,
                                        rel_appMerged_prevApps,
                                        rel_prevApps_POSCash,
                                        rel_prevApps_creditCard,
                                        rel_prevApps_installmentsPayments])

# View the entitySet 
entitySet
```

```
# aggregative primitives
aggPrims =  ["sum", "max", "min", "mean", "std", "count", "mode","skew"]

# DFS with specified primitives

# featureMatrix saved to csv, so this will be commented out for now, and the 
# reduced feature matrix can be loaded from csv when need be

featureMatrix, featureNames = ft.dfs(entityset = entitySet, target_entity = 'appMerged',
                       agg_primitives=aggPrims,
                       trans_primitives = [],
                       ignore_variables = {"creditCard" : ['SK_ID_CURR'],
                       "POSCash" : ['SK_ID_CURR'],
                       "installmentsPayments" : ['SK_ID_CURR'],
                       "appMerged" : ['set']},
                       max_depth=2)

print('%d Total Features' % len(featureNames))

# Save the featureMatrix
featureMatrix.to_csv('/content/drive/My Drive/CISC351/dataSet/preparedDataSet/featureMatrix.csv')
#featureMatrix.to_csv('/content/drive/My Drive/CISC351/dataSet/dataSet_OHE/featureMatrix_OHE.csv')
#featureMatrix.to_csv('/content/drive/My Drive/CISC351/dataSet/reducedDataSet/featureMatrix_reduced.csv')


# load the feature matrix
#featureMatrix = pd.read_csv('/content/drive/My Drive/CISC351/dataSet/reducedDataSet/featureMatrix_reduced.csv')
```

```
# Remove highly correlated variables.
# Highly correlated variables will be any variables that correlate with another with
# a greater than 0.8 Pearson Correlation coefficient

threshold = 0.8

featureCorrs = featureMatrix.corr().abs()
upperCorrs = featureCorrs.where(np.triu(np.ones(featureCorrs.shape), k=1).astype(np.bool))

count = 0
featureCorrsCols = list(featureCorrs.columns)
for col in featureCorrsCols:
  if any(upperCorrs[col] > threshold):
    featureMatrix.drop(col, axis=1, inplace=True)
    count += 1
    
print("There were %d collinear features dropped" % count)
    
```

```
print("The size of featureMatrix is %d by %d" % (featureMatrix.shape[0], featureMatrix.shape[1]))
```

```
# Determine if any of the features have near zero correlations with the target
threshold = 0.01 # Generous, but correlation isn't the be-all,end-all

featureCorrs = featureMatrix.corr().abs()
targetCorrelations = featureCorrs['TARGET'].dropna().sort_values() # dropna() (NOT inplace) b/c categoricals cant correlate
count = 0


featureNames = list(targetCorrelations.keys())

for feature in featureNames:
  if targetCorrelations[feature] < threshold:
    featureMatrix.drop(feature, axis=1, inplace=True)
    count += 1

print("There were %d insufficiently correlated features" % count)
```
